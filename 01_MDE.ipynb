```python
from scipy import stats

n = 60  # наблюдений на группу в неделю на основании исторических данных
mean = 93.18  # baseline RPM
std = 42.67
alpha = 0.05
beta = 0.20

# Обратный расчёт МДЕ при заданных n, α, β
df_init = n - 1
t_alpha = stats.t.ppf(1 - alpha/2, df_init)
t_beta = stats.t.ppf(1 - beta, df_init)

# Формула для двух выборок
mde_absolute = (t_alpha + t_beta) * std * (2 / n) ** 0.5
mde_relative = mde_absolute / mean * 100

print(f"Детектируемый МДЕ: {mde_relative:.1f}%")

```
